```python
#Створіть одновимірний масив (вектор) зі спадаючою послідовністю чисел, де кожне наступне число менше попереднього на 2, починаючи з 30, та виведіть його значення.
vector = [i for i in range(30, 0, -2)]
print(vector)
```

```python
#Створіть двовимірний масив (матрицю) розміром 3x3, заповніть його значеннями факторіалів натуральних чисел, починаючи з 1, та виведіть його значення.
import math
rows, cols = 3, 3
matrix = [[math.factorial(i + j) for j in range(cols)] for i in range(rows)]
for row in matrix:
    print(row)
```

```python
#Створіть масив розміром 5x5, заповніть його випадковими цілими числами від 1 до 10000 та виведіть його значення. 
import random
rows, cols = 5, 5
matrix = [[random.randint(1, 10000) for _ in range(cols)] for _ in range(rows)]
for row in matrix:
    print(row)
```

```python
#Створіть масив розміром 6x6, заповніть його випадковими дійсними числами в діапазоні від 0 до 1 та обчисліть суму всіх діагональних елементів. 
import numpy as np
rows, cols = 6, 6
matrix = np.random.rand(rows, cols)
print("Матриця:")
print(matrix)
diagonal_sum = np.trace(matrix)
print("\nСума діагональних елементів:", diagonal_sum)
```

```python
#Створіть два одновимірних масиви розміром 4, заповніть їх випадковими дійсними числами від -10 до 10 та виконайте на них поелементні операції додавання, віднімання та множення. 
import numpy as np
size = 4
array1 = np.random.uniform(-10, 10, size)
array2 = np.random.uniform(-10, 10, size)
print("Масив 1:", array1)
print("Масив 2:", array2)
addition_result = np.add(array1, array2)
print("\nРезультат додавання:", addition_result)
subtraction_result = np.subtract(array1, array2)
print("Результат віднімання:", subtraction_result)
multiplication_result = np.multiply(array1, array2)
print("Результат множення:", multiplication_result)
```

```python
#Створіть два вектори розміром 4, заповніть їх довільними дійсними числами та знайдіть їх скалярний добуток. 

import numpy as np
size = 4
vector1 = np.random.uniform(-10, 10, size)
vector2 = np.random.uniform(-10, 10, size)
print("Вектор 1:", vector1)
print("Вектор 2:", vector2)
dot_product = np.dot(vector1, vector2)

print("\nСкалярний добуток:", dot_product)
```

```python
#Створіть дві матриці розміром 2x3 та 3x3, заповніть їх випадковими цілими числами в діапазоні від 1 до 12 та перемножте їх між собою.
import numpy as np
rows1, cols1 = 2, 3
rows2, cols2 = 3, 3
matrix1 = np.random.randint(1, 13, size=(rows1, cols1))
matrix2 = np.random.randint(1, 13, size=(rows2, cols2))
print("Матриця 1:")
print(matrix1)

print("\nМатриця 2:")
print(matrix2)
product_matrix = np.dot(matrix1, matrix2)
print("\nРезультат перемноження:")
print(product_matrix)
```

```python
#Створіть матрицю розміром 3x3, заповніть її випадковими цілими числами в діапазоні від 1 до 7 та знайдіть її обернену матрицю. 
import numpy as np
rows, cols = 3, 3
matrix = np.random.randint(1, 8, size=(rows, cols))
print("Початкова матриця:")
print(matrix)
try:
    inverse_matrix = np.linalg.inv(matrix)
    print("\nОбернена матриця:")
    print(inverse_matrix)
except np.linalg.LinAlgError:
    print("\nОбернена матриця не існує, бо початкова матриця є особливою.")
```

```python
#Створіть матрицю розміром 5x5, заповніть її випадковими дійсними числами в діапазоні від -5 до 5 та транспонуйте її. 
import numpy as np
rows, cols = 5, 5
matrix = np.random.uniform(-5, 5, size=(rows, cols))
print("Початкова матриця:")
print(matrix)
transposed_matrix = np.transpose(matrix)
print("\nТранспонована матриця:")
print(transposed_matrix)
```

```python
#Створіть матрицю розміром 3x4 та вектор розміром 4, заповніть матрицю випадковими цілими числами в діапазоні від 1 до 10, а вектор - випадковими дійсними числами від 0 до 1, і знайдіть добуток матриці на вектор.
import numpy as np
rows, cols = 3, 4
vector_size = 4
matrix = np.random.randint(1, 11, size=(rows, cols))
vector = np.random.random(vector_size)
print("Матриця:")
print(matrix)
print("\nВектор:")
print(vector)
product_result = np.dot(matrix, vector)
print("\nРезультат добутку матриці на вектор:")
print(product_result)
```

```python
#Створіть матрицю розміром 2x3 та вектор розміром 3, заповніть матрицю випадковими дійсними числами від 0 до 1, а вектор - випадковими дійсними числами від 0 до 1, і знайдіть суму елементів вектору, які менше середнього значення вектору. 
import numpy as np
rows, cols = 2, 3
vector_size = 3
matrix = np.random.rand(rows, cols)
vector = np.random.rand(vector_size)
print("Матриця:")
print(matrix)
print("\nВектор:")
print(vector)
mean_value = np.mean(vector)
print("\nСереднє значення вектора:", mean_value)
filtered_elements = vector[vector < mean_value]
print("\nЕлементи вектору, які менше середнього значення:")
print(filtered_elements)
sum_of_filtered_elements = np.sum(filtered_elements)
print("\nСума елементів вектору, які менше середнього значення:", sum_of_filtered_elements)
```

```python
#Створіть дві матриці розміром 9x9, заповніть їх випадковими цілими числами в діапазоні від 1 до 10 та знайдіть відношення елементів другої матриці до першої (поелементне ділення). 
import numpy as np
rows, cols = 9, 9
matrix1 = np.random.randint(1, 11, size=(rows, cols))
matrix2 = np.random.randint(1, 11, size=(rows, cols))
print("Матриця 1:")
print(matrix1)
print("\nМатриця 2:")
print(matrix2)
result_matrix = np.divide(matrix2, matrix1)
print("\nВідношення елементів другої матриці до першої:")
print(result_matrix)
```

```python
#Створіть матрицю розміром 6x6, заповніть її випадковими цілими числами в діапазоні від -50 до 50 та знайдіть середнє арифметичне значення елементів матриці. 
import numpy as np
rows, cols = 6, 6
matrix = np.random.randint(-50, 51, size=(rows, cols))
print("Матриця:")
print(matrix)
average_value = np.mean(matrix)
print("\nСереднє арифметичне значення елементів матриці:", average_value)
```

```python
#Створіть матрицю розміром 10x10, заповніть її випадковими дійсними числами в діапазоні від -2 до 2 та знайдіть вектор-стовпчик, що містить найменше за модулем значення елементу у кожному рядку матриці.
import numpy as np
rows, cols = 10, 10
matrix = np.random.uniform(-2, 2, size=(rows, cols))
print("Матриця:")
print(matrix)
min_abs_values_vector = np.argmin(np.abs(matrix), axis=1)
result_vector = matrix[np.arange(rows), min_abs_values_vector][:, np.newaxis]
print("\nВектор-стовпчик з найменшими за модулем значеннями у кожному рядку:")
print(result_vector)
```
